```python
import numpy as np
import pandas as pd
from sklearn import svm, metrics, tree
import matplotlib.pyplot as plt
from IPython.display import display
from astropy.table import Table
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_curve
from sklearn import preprocessing
from sklearn.metrics import precision_recall_fscore_support
from sklearn.svm import SVC  
```

```python
df= pd.read_csv('dadosProvas2016.csv')
dfAlunosTeste = pd.read_csv('dadosProvas2017.csv')
dfMedias = pd.DataFrame()
```

```python
dfAlunosTeste[0:10]
```

```python
dfAlunoAprovacao = dfAlunosTeste.groupby(['Aluno'],as_index=False).agg({'Nota':'mean'})
for index, row in dfAlunoAprovacao.iterrows():
    if(row['Nota'] >= 5.0):
        dfAlunoAprovacao.set_value(index,'Nota',1)
    else:
        dfAlunoAprovacao.set_value(index,'Nota',0)

dfAlunoAprovacao[['Nota']] = dfAlunoAprovacao[['Nota']].astype(int)

for index, row in dfAlunosTeste.iterrows():
    aluno = int(row['Aluno'])
    aprovado = dfAlunoAprovacao.loc[dfAlunoAprovacao['Aluno'] == aluno]['Nota'].values[0]
    dfAlunosTeste.set_value(index,'Aprovacao',aprovado)
    
    
dfAlunosTeste[['Aprovacao']] = dfAlunosTeste[['Aprovacao']].astype(int)

for index, row in dfAlunosTeste.iterrows():
    if(row['Nota'] >= 5.0):
        dfAlunosTeste.set_value(index,'Nota',1)
    else:
        dfAlunosTeste.set_value(index,'Nota',0)

dfAlunosTeste[['Nota']] = dfAlunosTeste[['Nota']].astype(int)
```

```python
dfAlunosTeste[0:10]
```

```python
def casting(df,arrayAprovacao):
    df[['Nota']] = df[['Nota']].astype(int)
    df['Duracao'] = df['Duracao'].astype(int)
    df['Qntd_Linhas'] = df['Qntd_Linhas'].astype(int)
    df['Qntd_Linhas_Filtrado'] = df['Qntd_Linhas_Filtrado'].astype(int)
    df['Testes'] = df['Testes'].astype(int)
    df['Submissoes'] = df['Submissoes'].astype(int)
    df['PP_Paste'] = df['PP_Paste'].astype(float)
    df['PP_Delete'] = df['PP_Delete'].astype(float)
    df['Linhas_Log'] = df['Linhas_Log'].astype(int)
    df['Aprovacao'] = arrayAprovacao
    df['Aprovacao'] = df['Aprovacao'].astype(int)
    return df
```

```python
def criaDFAlunoMedia(ceil,aluno,dfPassado,dfFinal):
    dfAluno1 = dfPassado.loc[dfPassado['Aluno'] == aluno][0:ceil]
    dfAluno1 = dfAluno1.groupby(['Aluno'],as_index=False).agg({'Duracao': 'mean', 'Qntd_Linhas': 'mean', 'Qntd_Linhas_Filtrado': 'mean'
                                                 , 'Testes': 'mean', 'Submissoes': 'mean', 'PP_Paste': 'mean'
                                                 , 'PP_Delete': 'mean', 'Linhas_Log': 'mean', 'Nota': 'sum'})
    dfFinal = dfFinal.append(dfAluno1, ignore_index = True)
    return dfFinal
```

```python
#Para validacao vou ter um vetor que vai armazenar os valores por iteracao
iteracoes = np.arange(1,8)
train_accuracy =np.empty(len(iteracoes))
test_accuracy = np.empty(len(iteracoes))
mcc_accuracy = np.empty(len(iteracoes))
f1_accuracy = np.empty(len(iteracoes))
```

```python
for qntdProvas in range(1,7):
    print(">Coletando dados de",qntdProvas,"provas\n")
    dfMedias = pd.DataFrame()
    alunoAnterior = 0
    for index, row in dfAlunosTeste.iterrows():
        alunoSelecionado = int(row['Aluno'])
        if(alunoSelecionado != alunoAnterior):
            contador = 1
            alunoAnterior = alunoSelecionado
            dfMedias = criaDFAlunoMedia(qntdProvas,alunoSelecionado,dfAlunosTeste,dfMedias)
    print(">Coletei",index,"linhas do dataframe. Iniciando classificacao....")
    arrayAprovacao = dfAlunoAprovacao['Nota'].values
    dfMedias = casting(dfMedias,arrayAprovacao)
    dfMedias['is_train'] = np.random.uniform(0, 1, len(dfMedias)) <= .75
    treino, teste = dfMedias[dfMedias['is_train']==True], dfMedias[dfMedias['is_train']==False]
    print('Quantidade de dados de treino:', len(treino))
    print('Quantidade de dados de teste:',len(teste))
    features = dfMedias.columns[1:10]
    y = treino['Aprovacao']
    
    clf = RandomForestClassifier(max_depth=2, random_state=0)
    clf.fit(treino[features], y)
    y_pred = clf.predict(teste[features])
    
    mcc = matthews_corrcoef(teste['Aprovacao'], y_pred)
    mcc_accuracy[qntdProvas] = mcc
    
    f1 = precision_recall_fscore_support(teste['Aprovacao'], y_pred, average='binary')
    #f1_accuracy[qntdProvas] = f1
    
    score = clf.score(treino[features], y)
    train_accuracy[qntdProvas] = score
    
    scoreTest = clf.score(teste[features], teste['Aprovacao'])
    test_accuracy[qntdProvas] = scoreTest
    
    #matriz de confusao
    matrizConfu = confusion_matrix(teste['Aprovacao'], y_pred)
    #importancia de cada atributo
    importancia = list(zip(treino[features], clf.feature_importances_))

    print("\n>Resultados da coleta de",qntdProvas,"prova:\nScore: ",score,"\nMCC: ",mcc,"\nF1: ",f1,"\nMatriz de confusao:\n")
    print(matrizConfu)
    print("\nCalculo de relevancia de cada atributo:",importancia,"\n**********\n")
    #print(dfMedias[0:100])
    
```

```python
brita = np.arange(1,7)
test_accuracy = np.delete(test_accuracy,0)
train_accuracy = np.delete(train_accuracy,0)
#Generate plot
plt.title('Random Forest variando com o número de provas')
plt.plot(brita, test_accuracy, label='Acurácia de teste')
plt.plot(brita, train_accuracy, label='Acurácia de treino')
plt.legend()
plt.xlabel('Número de provas')
plt.ylabel('Acurácia')
plt.show()
```

```python
brita = np.arange(1,7)
mcc_accuracy = np.delete(mcc_accuracy,0)
#Generate plot
plt.title('MCC variando com número de provas')
plt.plot(brita, mcc_accuracy, label='MCC')
plt.legend()
plt.xlabel('Número de provas')
plt.ylabel('Pontuação')
plt.show()
```
